```python
import matplotlib.pyplot as plt
import seaborn as sns

# New gene and condition names
genes = ['TP53', 'BRCA1', 'EGFR', 'VEGFA', 'MYC']
conditions = ['Normal', 'Hypoxia', 'Drug_A', 'Drug_B']

# New gene expression data
gene_expr_data = [
    [0.65, 0.80, 0.45, 0.50],
    [0.92, 0.75, 0.60, 0.70],
    [0.55, 0.40, 0.88, 0.76],
    [0.45, 0.60, 0.70, 0.95],
    [0.82, 0.90, 0.78, 0.65]
]

# Plotting
plt.figure(figsize=(8, 6))
sns.heatmap(gene_expr_data, annot=True, fmt=".2f", cmap="Oranges",
            xticklabels=conditions, yticklabels=genes)
plt.title('Gene Expression Heatmap (New Data)', weight='bold')
plt.xlabel('Conditions', weight='bold')
plt.ylabel('Genes', weight='bold')
plt.tight_layout()
plt.savefig('heatmap_new.png', dpi=300)
plt.show()
```
